```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Admission_Predict.csv")
```

```python
# To make chance of admit column in binary form o/p
from sklearn.preprocessing import Binarizer
bi=Binarizer(threshold=0.75)
df['Chance of Admit ']=bi.fit_transform(df[['Chance of Admit ']])
```

```python
df
```

```python
df.head()
```

```python
x = df.drop("Chance of Admit ", axis = 1)
y = df["Chance of Admit "]
```

```python
x
```

```python
y
```

```python
y = y.astype("int")
```

```python
y
```

```python
sns.countplot(x = y);
```

```python
y.value_counts()
```

```python
#CROSS VALIDATION 
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train, y_test= train_test_split(x,y,random_state=0,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
# import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(random_state=0)
```

```python
classifier
```

```python
# To Train the model using classifire
classifier.fit(x_train,y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
result=pd.DataFrame({
    'Actual':y_test,
    'predict':y_predict
})
```

```python
result
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
from sklearn.metrics import classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_predict,y_test)
```

```python
# for finding accuracy, precision, recall, f1 score 
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
precision_score(y_test,y_predict)
```

```python
recall_score(y_test,y_predict)
```

```python
f1_score(y_test,y_predict)
```

```python
# for whole classification report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```
